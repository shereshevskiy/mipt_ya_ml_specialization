```python
import requests
import bs4
```

### распарсите: а) из статьи https://en.wikipedia.org/wiki/Bias-variance_tradeoff все заголовки верхнего уровня;

```python
req = requests.get('https://en.wikipedia.org/wiki/Bias-variance_tradeoff')
soup = bs4.BeautifulSoup(req.text, 'lxml')
print(soup.prettify())
```

```python
for p in soup.findAll('li', attrs={'class': "toclevel-1"}):
    p_text = p.find('span', attrs={'class':"toctext"}).text
    p_number = p.find('span', attrs={'class':"tocnumber"}).text
    print ('{} {}'.format(p_number, p_text))
```
