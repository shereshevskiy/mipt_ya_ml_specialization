```python
import requests
import bs4
```

```python
url ="https://en.wikipedia.org/wiki/Bias%E2%80%93variance_tradeoff"
url2 = "https://en.wikipedia.org/wiki/Category:Machine_learning_algorithms"
```

### Part 1

https://en.wikipedia.org/wiki/Bias%E2%80%93variance_tradeoff

```python
req = requests.get(url)
req.status_code
```

```python
type(req)
```

```python
print(req.text)
```

```python
parser = bs4.BeautifulSoup(req.text,'lxml')
```

```python
y = parser.find_all('h1')
for result in y:
    print (result.text)
    print ("\n------\n")
```

### Part 2
https://en.wikipedia.org/wiki/Category:Machine_learning_algorithms

```python
req2 = requests.get(url2)
req2.status_code
```

```python
parser2 = bs4.BeautifulSoup(req2.text, 'lxml')
```

```python
t = parser2.findAll(lambda tag: tag.name == 'a' and tag.findParent('li') and \
                   tag.findParent('li').findParent('ul') and \
                   tag.findParent('li').findParent('ul').findParent('div', attrs={'class':'mw-category-group'}))
for result in t:
    print(result.text)
    print ("\n------\n")
```
